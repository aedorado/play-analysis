```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn
seaborn.set() #make the plots look pretty
```

```python
def get_file_lines(filename):
    with open(filename) as f:
        filelines = list(f)
    return filelines
```

```python
clines = get_file_lines('compiled_info_centrality')

def rating_vs_connectors(filelines):
    clblines = [x for x in filelines if ':' in x and '::' not in x]
    x = []
    y = []
    for line in clblines:
        val = [int(s) for s in line.split() if s.isdigit()]
        x.append(val[0])
        y.append(val[1])
    y_pos = np.arange(len(x))
    print x, y
    
    plt.bar(y_pos, y, align='center', alpha=0.5)
    plt.xticks(y_pos, x)
    plt.xlabel('# Top Rated Apps')
    plt.ylabel('# Connectors and Connected')
    plt.title('Rating vs Connectors')
    mpl.rcParams['xtick.labelsize'] = 10
    mpl.rcParams['ytick.labelsize'] = 10

    plt.show()
    
rating_vs_connectors(clines)
```

```python
topline = [x for x in filelines if 'top' in x]
tops = []
outs = []
lens = []
for line in topline[:len(topline) / 4]:
    val = [int(s) for s in line.split() if s.isdigit()]
    tops.append(val[0])
    outs.append(val[1])
    lens.append(val[2])
print len(tops), (tops), outs, lens
```

```python
"""
========
Barchart
========

A bar plot with errorbars and height labels on individual bars
"""

N = len(outs)
men_means = outs
men_std = lens

ind = np.arange(N)  # the x locations for the groups
width = 0.25       # the width of the bars

fig, ax = plt.subplots(figsize=(35, 15))
rects1 = ax.bar(ind, men_means, width, color='r')

women_means = lens
women_std = lens
rects2 = ax.bar(ind + width, women_means, width, color='y')

# add some text for labels, title and axes ticks
ax.set_ylabel('Number', fontsize=20)
ax.set_title('Scores by group and Centrality', fontsize=20)
ax.set_xticks(ind + width / 2)
ax.set_xticklabels([(x + 6) / 6 * 10 for x in range(0, len(outs))])
mpl.rcParams['xtick.labelsize'] = 20
mpl.rcParams['ytick.labelsize'] = 20

ax.legend((rects1[0], rects2[0]), ('Outdegree', 'Centrality'))


def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/1.5, 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom', fontsize=20)

autolabel(rects1)
autolabel(rects2)

plt.show()
```

```python
def plot_from_map(mymap, xlabel='', ylabel='', title=''):
    x = []
    y = []
    for key in sorted(mymap.iterkeys()):
        if key is not None:
            x.append(key), y.append(mymap[key])

    y_pos = np.arange(len(x))
    print x, y

    plt.bar(y_pos, y, align='center', alpha=0.5, width=0.5)
    plt.xticks(y_pos, x)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    mpl.rcParams['xtick.labelsize'] = 7
    mpl.rcParams['ytick.labelsize'] = 10

    plt.show()
```
